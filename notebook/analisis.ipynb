```python
import pandas as pd

paises = pd.read_csv('../data/Paises.csv')
poblacion = pd.read_csv('../data/Poblacion.csv')
mortalidad = pd.read_csv('../data/Mortalidad-Infantil.csv')
esperanza = pd.read_csv('../data/Esperanza-de-Vida.csv')
```

```python
paises.drop_duplicates(inplace=True)
poblacion.drop_duplicates(inplace=True)
mortalidad.drop_duplicates(inplace=True)
esperanza.drop_duplicates(inplace=True)
```

```python
print(paises.isnull().sum())
print(poblacion.isnull().sum())
print(mortalidad.isnull().sum())
print(esperanza.isnull().sum())
```

```python
combinado = (
    paises
    .merge(poblacion, on='País', how='inner')
    .merge(mortalidad, on='País', how='inner')
    .merge(esperanza, on='País', how='inner')
)
combinado.info()
```

```python
combinado.describe()
```

```python
combinado.select_dtypes(include='number').corr()
```

```python
total_mundial = combinado['Población'].sum()
print("Población total mundial:", total_mundial)
```

```python
continentes = (
    combinado
    .groupby('Continentes', as_index=False)
    .agg(Poblacion=('Población', 'sum'))
    .sort_values('Poblacion', ascending=False)
)
print("\nPoblación por continente:")
print(continentes)
```

```python
continentes['Porcentaje_Poblacion']=continentes['Poblacion']/continentes['Poblacion'].sum()*100
continentes
```

```python
top_5 = combinado.sort_values('Población', ascending=False).head(5)

print("\nTop 5 países más poblados:")
print(top_5[['País', 'Población']])
```
